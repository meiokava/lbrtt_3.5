<font size="4">Визуализация взаимосвязи между двумя переменными в линейной функции спроса.
1. Предположим, у нас есть функция спроса на продукт:
$$Q = a - bP$$

    где Q - требуемое количество,
    
    P - цена продукта,
    
    а a и b - константы.
    
    
2. Будем использовать эту функцию спроса для анализа влияния изменений цены на количество требуемого продукта.
В этой задаче определим функцию спроса, используя уравнение для линейной функции спроса.
    
    
3. Генерируется значения цен, с помощью использования функции linspace от numpy, график функции спроса строится с помощью plt.plot.

    
4. Границы и метки графика, устанавливаются с помощью plt.xlim, plt.ylim, plt.xlabel и plt.ylabel. Заголовок к графику добавляется при использовании plt.title.

Наконец, добавляется аннотацию к графику, используя plt.annotate, чтобы указать конкретную точку на графике, где цена = 10, а требуемое количество = 50.

На графики наклон линии отражает ценовую эластичность спроса, в то время как пересечение представляет максимальное количество, требуемое по нулевой цене.

```python
import numpy as np
import matplotlib.pyplot as plt

# define the demand function
def demand_func(price):
    a = 100
    b = 5
    return a - b*price

# generate price values
price = np.linspace(0, 20, 100)

# plot the demand function
plt.plot(price, demand_func(price), label='Demand Function')

# set the limits and labels
plt.xlim(0, 20)
plt.ylim(0, 120)
plt.xlabel('Price')
plt.ylabel('Quantity Demanded')
plt.title('Linear Demand Function')

# add axes annotations
plt.annotate('Price = 10, Quantity Demanded = 50', xy=(10, 50), xytext=(5, 80),
             arrowprops=dict(facecolor='black', shrink=0.05))

# add legend and grid
plt.legend()
plt.grid()

# show the plot
plt.show()
```
