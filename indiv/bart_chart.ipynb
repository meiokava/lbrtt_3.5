<font size="4">Имеется набор данных, содержащий оценки 50 учащихся класса на выпускном экзамене. Нам необходимо визуализировать распределение баллов, используя столбчатую диаграмму с ячейками фиксированной ширины.

1. Сначала создается список баллов для 50 студентов. Затем устанавливается количество ячеек равным 10 и вычисляется ширину ячейки как диапазон оценок, деленный на количество ячеек.

2. Используется функцию arange от numpy для создания массива ячеек с минимальными и максимальными значениями в качестве конечных точек и шириной ячейки в качестве размера шага.Затем используется plt.hist для создания гистограммы с оценками в качестве данных и ячейками в качестве краев.

3. Установливается цвет краев полос на черный, а ширину линии - на 1,2 для лучшей видимости.

4. Задается заголовок и метки графика, используя plt.title, plt.xlabel и plt.ylabel.

Результирующая столбчатая диаграмма отображает распределение итоговых результатов экзамена с 10 ячейками фиксированной ширины. По оси x отображаются значения баллов, а по оси y - частота использования каждой ячейки баллов.    

```python
import matplotlib.pyplot as plt
import numpy as np

# create data
scores = [63, 75, 82, 67, 91, 78, 81, 86, 88, 69, 78, 92, 85, 76, 80, 75, 
          72, 90, 79, 83, 89, 81, 73, 77, 84, 72, 77, 86, 71, 68, 70, 81, 
          92, 86, 88, 79, 75, 73, 83, 80, 77, 90, 84, 81, 79, 85, 87, 71, 76]

# set the number of bins and bin width
num_bins = 10
bin_width = (max(scores) - min(scores)) / num_bins

# create the bins
bins = np.arange(min(scores), max(scores) + bin_width, bin_width)

# create the histogram
plt.hist(scores, bins=bins, edgecolor='black', linewidth=1.2)

# set the title and labels
plt.title('Distribution of Final Exam Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')

# show the plot
plt.show()
```
