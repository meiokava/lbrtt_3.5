<font size="4"> Имеется набор данных, содержащий распределение ежемесячного дохода семьи по различным категориям расходов. Нам нужно визуализировать долю дохода, распределенную по каждой категории, с помощью круговой диаграммы.
1. Сначала создаются два списка, один для категорий и один для соответствующих расходов. Затем используется plt.pie для создания круговой диаграммы с расходами в качестве данных и категориями в качестве меток. 

2. Используется параметр autopct для форматирования процентных значений по каждой категории.

3. Заголовок графика создается с помощью plt.title.

4. Используется plt.show, для отображения круговую диаграмму.

Сама круговая диаграмма отображает шесть срезов, представляющих долю ежемесячного дохода, распределенную по каждой категории расходов. Надписи на каждом срезе соответствуют шести категориям, а размер каждого среза соответствует проценту дохода, распределяемого на эту категорию.

```python
import matplotlib.pyplot as plt

# create data
categories = ['Housing', 'Food', 'Transportation', 'Utilities', 'Entertainment', 'Other']
expenses = [1200, 400, 300, 200, 100, 100]

# create the pie chart
plt.pie(expenses, labels=categories, autopct='%1.1f%%')

# set the title
plt.title('Monthly Expenses')

# show the plot
plt.show()
```
